```python
dfteam=df[df['target']=='All NBA TEAM']
```

```python
seasons_=['2007-08','2008-09','2009-10','2010-11','2011-12','2012-13','2013-14','2014-15','2015-16','2016-17','2017-18','2018-19','2019-20']
```

```python
dfteam
```

```python
dfpf=dfteam.loc[(dfteam['POS'] == 'PF')]
dfsf=dfteam.loc[(dfteam['POS'] == 'SF')]
dfsg=dfteam.loc[(dfteam['POS'] == 'SG')]
dfpg=dfteam.loc[(dfteam['POS'] == 'PG')]
dfc=dfteam.loc[(dfteam['POS'] == 'C')]
```

```python
dfpositional=pd.concat([dfpf,dfsf,dfsg,dfpg,dfc])
```

```python
dfpositional
```

```python
dfpositional['POS'].value_counts(normalize=True)
```

```python
#Analysis to see most popular genres produced.
counts_ = dfpositional['POS'].value_counts(normalize=True)

colors=['steelblue'] *22
colors[0:5]=['lightcoral']*5

fig, ax = plt.subplots(figsize=(15, 6))

plt.title('All-NBA Team by Position',fontsize=18)
plt.xlabel('Position')
plt.ylabel('PErcentage of Players')

ax.bar(x=counts_.index, height=counts_,color=colors)
ax.set_xticklabels(labels=counts_.index, rotation=80);
```
